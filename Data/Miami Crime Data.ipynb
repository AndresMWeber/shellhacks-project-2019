```python
import pandas as pd
import numpy as np 
from pygeocoder import Geocoder 
from geopy.geocoders import Nominatim
import json






```

```python
miami_crime=pd.read_csv('Miami-crime-data.csv')
miami_crime.head()
```

```python
miami_crime.dtypes
```

```python
# Dropping rows in Description that contains "OVERDUE RENTAL"

droping=miami_crime[miami_crime['Description'] == 'OVERDUE RENTAL'].index
droping1=miami_crime[miami_crime['Description']=='FALSE REPORT OF COMMISSION OF CRIME(S)'].index
droping2=miami_crime[miami_crime['Description']=='OFFERING FOR OR ENGAGING IN PROSTITUTION'].index
miami_crime.drop(droping, inplace=True)
miami_crime.drop(droping1, inplace=True)
miami_crime.drop(droping2, inplace=True)


```

```python
# Categorizing Description Column into 5 categories

category1=['MOTOR','VEHICLE']
category2=['THEFT','LARCENY']
category3=['BURGLARY']
category4=['ASSAULT','MURDER','HOMICIDE','MOLESTATION','THREATENS','LASCIVIOUS']
category5=['BATTERY']
category6=['ROBBERY','$']
category7=['VANDALISM','ARSON']
category8=['FRAUD','FRAUDULENT','FRAUDULENTLY']
category10=['DRUGS/ALCOHOL VIOLATIONS','CANNABIS','COCAINE','ALCOHOL','SUBSTANCE','POSSESSION','POSSESS','DRUG','DRUGS','DUI']
category11=['WEAPONS','FIREARM']

for x in miami_crime['Description']:
    if category1[0] in x or category1[1] in x:
        miami_crime['Description']=miami_crime['Description'].replace(x,'Motor Vehicle Theft')
        
    elif category2[0] in x or category2[1] in x:
        miami_crime['Description']=miami_crime['Description'].replace(x, 'Theft/Larceny')
        
    elif category3[0] in x:
        miami_crime['Description']=miami_crime['Description'].replace(x,'Burglary')
        
    elif category4[0] in x or category4[1] in x or category4[2] in x or category4[3] in x or category4[4] in x or category4[5] in x:
        miami_crime['Description']=miami_crime['Description'].replace(x,'Assault') 
        
    elif category5[0] in x:
        miami_crime['Description']=miami_crime['Description'].replace(x,'Battery') 
        
    elif category6[0] in x or category6[1] in x:
        miami_crime['Description']=miami_crime['Description'].replace(x,'Robbery') 
    
    elif category7[0] in x or category7[1] in x:
        miami_crime['Description']=miami_crime['Description'].replace(x, 'Vandalism')
      
    elif category8[0] in x or category8[1] in x or category8[2] in x:
        miami_crime['Description']=miami_crime['Description'].replace(x, 'Fraud') 
        
    elif category10[0] in x or category10[1] in x or category10[2] in x or category10[3] in x or category10[4] in x or category10[5] in x or category10[6] in x or category10[7] in x or category10[8] in x or category10[9] in x:
        miami_crime['Description']=miami_crime['Description'].replace(x, 'Drugs/Alcohol Violations')   
        
    elif category11[0] in x or category11[1] in x:
        miami_crime['Description']=miami_crime['Description'].replace(x, 'Weapon Charges')   

miami_crime
```

```python
# Removing Block from the locations

remove='BLOCK'

miami_crime['Location']=[x.replace(remove,'') for x in miami_crime['Location']]
        
miami_crime
```

```python
# Removing '&','/' from Location
miami_crime['Location']=[x.split('&')[0] for x in miami_crime['Location']]
miami_crime['Location']=[x.split('/')[0] for x in miami_crime['Location']]

miami_crime
```

```python
# Adding Miami to the end of each string
miami_crime['Location']=[x+ ' Miami' for x in miami_crime['Location']]
miami_crime
```
